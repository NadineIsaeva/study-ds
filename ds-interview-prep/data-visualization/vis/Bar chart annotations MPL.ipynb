```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import os
#os.listdir()

pd.set_option('display.max_row', 1000)
pd.set_option('display.max_columns', 50)

data = pd.read_csv('data/5kings_battles_v1.csv')
```

```python
# Horizontal bar chart

ax = data['region'].value_counts().plot(kind='barh', figsize=(10,7),
                                  color='coral', fontsize=13)

ax.set_alpha(0.8)
ax.set_title('Where were the battles fought?', fontsize=18)
ax.set_xlabel('Number of battles', fontsize=18)
ax.set_xticks([0, 5, 10, 15, 20])

# Create list to collect plt.patches data
totals = []

for i in ax.patches: # 7 patches - as bars
    totals.append(i.get_width()) # get values for each bar
    
total = sum(totals) # overall quantity of battles

for i in ax.patches:
    ax.text(i.get_width() + 0.3, # padding from right
           i.get_y() + 0.38, # padding from top
           str(round((i.get_width()/total * 100), 2)) + '%',
           fontsize=15,
           color='dimgrey')
# largest on top
ax.invert_yaxis()
```

```python
# Vertical bar chart

ax = data['region'].value_counts().plot(kind='bar', figsize=(10,7),
                                       color='coral', fontsize=13)
ax.set_alpha(0.8)
ax.set_title('Where were the battles fought?', fontsize=18)
ax.set_xlabel('Number of Battles', fontsize=18)
ax.set_yticks([0, 5, 10, 15, 20])

totals = []

for i in ax.patches:
    totals.append(i.get_height()) # get values for each bar
    
total = sum(totals)

for i in ax.patches:
    ax.text(i.get_x() - 0.03, # right padding
           i.get_height() + 0.5, # bottom padding
           str(round((i.get_height()/total)*100, 2)) + '%',
           fontsize=15,
           color='dimgrey')
```

```python
# Vertical bar chart with text inside the bars

ax = data['attacker_outcome'].value_counts().plot(kind='bar', figsize=(10, 7),
                                                 color='indigo', fontsize=13)
ax.set_alpha(0.8)
ax.set_title("Do attackers usually win or loose?", fontsize=18)
ax.set_ylabel("Number of Battles", fontsize=18);
ax.set_yticks([0, 5, 10, 15, 20, 25, 30, 35, 40])

totals = []

for i in ax.patches:
    totals.append(i.get_height())
    
total = sum(totals)

for i in ax.patches:
    ax.text(i.get_x() + 0.12, #left padding
           i.get_height() - 3, # top padding
           str(round((i.get_height()/total)*100, 2)) + '%',
           fontsize=22,
           color='white' 
           )
```

```python
# Horizontal bar chart with counts

losses = data[data['attacker_outcome'].str.contains('loss', na=False)]

ax = losses['attacker_king'].value_counts().plot(kind='barh', figsize=(10,7),
                                                color='blue', fontsize=13)
ax.set_alpha(0.8)
ax.set_title('Who were the attackers who lost?', fontsize=18)
ax.set_xlabel('Number of Battles', fontsize=18)
ax.set_xticks([0, 1, 2, 3, 4, 5])

for i in ax.patches:
    ax.text(i.get_width() + 0.1,
           i.get_y() + 0.31,
           str(i.get_width()),
              fontsize=15,
              color='dimgrey')
    
ax.invert_yaxis()
```

```python
# Vertical bar plot with multiple columns

ax = losses[['attacker_size', 'defender_size']].plot(kind='bar', figsize=(15,7),
                                               color=['blue', 'grey'], fontsize=13)
ax.set_alpha(0.8)
ax.set_title('For attacker losses, what was the difference in size?', fontsize=18)
ax.set_ylabel('Number of Troops', fontsize=18)
ax.set_yticks([0, 20000, 40000, 60000, 80000, 100000, 120000, 140000])
ax.set_xticklabels(["Robb v Joff/Tommen", "Joff/Tommen v Robb", 
                    "Stannis v Joff/Tommen", "Robb v Joff/Tommen", 
                    "Stannis v Mance"], rotation=0, fontsize=11)

for i in ax.patches: # 10 rectangles
    ax.text(i.get_x(),
           i.get_height(),
           str(i.get_height()),
           fontsize=11,
           color='dimgrey',
           rotation=45)
```

```python
# Horizontal bar chart with multiple columns

ax = losses[['attacker_size', 'defender_size']].plot(kind='barh', figsize=(10,7),
                                                    color=['blue', 'grey'], fontsize=13)
ax.set_alpha(0.8)
ax.set_title('For attacker losses, what was the difference in size?', fontsize=18)
ax.set_ylabel('First name is attacker', fontsize=18)
ax.set_xticks([0, 20000, 40000, 60000, 80000, 100000, 120000, 140000])
ax.set_yticklabels(["Robb v Joff/Tommen", "Joff/Tommen v Robb", 
                    "Stannis v Joff/Tommen", "Robb v Joff/Tommen", 
                    "Stannis v Mance"])

for i in ax.patches:
    ax.text(i.get_width() + 700,
           i.get_y() + 0.18,
           str(i.get_width()),
           fontsize=11,
           color='dimgrey')
    
ax.invert_yaxis()
```
